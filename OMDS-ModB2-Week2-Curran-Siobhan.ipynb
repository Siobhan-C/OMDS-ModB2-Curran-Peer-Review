<b>Question: Encode any categorical variables (e.g. with one-hot encoding.)</b>

This task is similar to last week's homework (#9), where we one-hot encoded the AccountType column. In that task, we created new columns for "checking," "savings," and "cd" so that machine learning algorithms could work with the data. For this task, I googled "one-hot encoded columns" and found two helpful resources:

According to Geeksfor Geeks, "One Hot Encoding is a method for converting categorical variables into a binary format. It creates new columns for each category where 1 means the category is present and 0 means it is not." This process is useful because machine learning algorithms work better with numerical data, and one-hot encoding allows categorical data to be represented numerically.

I started by applying one-hot encoding to the neighbourhood_group column, assuming it was categorical. However, upon further review of the dataset (including checking the actual CSV file), I found that neighbourhood_group only contains NaN values (which were updated to 'Unknown'). As a result, there were no other categories beyond NaN/unknown, and only a single column—neighbourhood_group_Unknown—was created. (note: one-hot endcoding would not work with the missing vales. Due to this, I populated the neighbourhood_group with data of  unkonwn).

Since the neighbourhood_group column only had one value (i.e., Unknown), the one-hot encoding resulted in just one column. In this case, Unknown represents all the data, and I don’t need further encoding for this particular column.

```python
#update missing values in the 'neighbourhood_group' column with 'Unknown'
inside_airbnb['neighbourhood_group'] = inside_airbnb['neighbourhood_group'].fillna('Unknown')
```

```python
#apply one-hot encoding to the 'neighbourhood_group' column. 
encoded_neighbourhood_group = pd.get_dummies(inside_airbnb['neighbourhood_group'], prefix='neighbourhood_group', drop_first=False)

#combining the one-hot encoded columns back to the original DF
inside_airbnb_encoded = pd.concat([inside_airbnb, encoded_neighbourhood_group], axis=1)
print(inside_airbnb_encoded[['neighbourhood_group', 'neighbourhood_group_Unknown']].head())
```
