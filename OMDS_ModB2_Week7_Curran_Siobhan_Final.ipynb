**Answer:** How could you use preattentive attributes to do that? Suggest several possibilities.

There are several ways to highlight the last three months of data, such as:
1.   Highlight the recent three months using a high-contrast color while keeping older data in muted tones.
2. Emphasize the latest data by increasing marker size, line thickness, or by using distinctive shapes compared to earlier data points.
3. Differentiate the time period by spatially grouping or framing it with subtle background highlights or borders.
4. Add motion or annotations to draw the reviewer's attention to the specific timeframe we are trying to highlight (i.e., the last three months).

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

I recreated the heatmap from "Preattentive Attributes in Graphs  Color" (p. 119) with a fun St. Patrick's Day twist to celebrate my Irish roots. Enjoy!   


```python
import matplotlib.pyplot as plt

#fake/created data: rows = Irish counties, columns = types of St. Patrick's Day events.
data = np.array([
    [45, 70, 55],  # Dublin
    [30, 60, 40],  # Cork
    [25, 50, 35],  # Galway
    [20, 45, 30],  # Limerick
    [35, 65, 50]   # Belfast
])

#create df with data
counties = ['Dublin', 'Cork', 'Galway', 'Limerick', 'Belfast']
events = ['Parades', 'Pub Events', 'Festivals']
df = pd.DataFrame(data, index=counties, columns=events)
plt.figure(figsize=(10, 8))

#addd title
plt.suptitle("Happy St. Patrick's Day! 03/17!", fontsize=24, color='green', fontweight='bold', y=0.98)

#create heat map w/green-hues on colormap
plt.imshow(df, cmap='Greens', interpolation='nearest')
cbar = plt.colorbar(label="Number of Shamrockin' Events")
plt.xticks(np.arange(len(events)), events, rotation=45, color='darkgreen', fontsize=12)
plt.yticks(np.arange(len(counties)), counties, color='darkgreen', fontsize=12)

plt.title(" St. Paddy's Day Heatmap: Luck & Shenanigans ", fontsize=16, color='darkgreen', pad=20)
for i in range(len(counties)):
    for j in range(len(events)):
        plt.text(j, i, f" {df.iloc[i, j]}", ha='center', va='center', color='white', fontsize=12)

```
