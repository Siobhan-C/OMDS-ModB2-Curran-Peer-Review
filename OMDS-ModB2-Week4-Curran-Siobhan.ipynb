Both histograms and boxplots have distinct advantages and limitations in representing data.

<b><i>Histograms - </i></b>

Advantages:
- Histograms offer a straightforward way to visualize the frequency of values within specified ranges, making it easy to quickly understand how data points are distributed.
- They provide a clear depiction of how the data is spread across different price ranges.
- Histograms help identify the concentration of data points, as well as the distribution's tails, which can be useful for spotting patterns and trends.

Limitations:
- The number of bins chosen for the histogram can influence interpretation. Too many bins may create unnecessary complexity, while too few can fail to capture the variation within the data.
- Histograms do not clearly highlight extreme outliers, which can be important for understanding the presence of unusually high or low values, especially in pricing data.

<b><i>Boxplot - </i></b>

Advantages:
- Boxplots deliver an efficient summary of the data, providing insight into the central tendency, spread, and skewness, including the median and quartiles.
- They are highly effective for comparing different sets of data, making them useful when analyzing multiple distributions side-by-side.
- The boxplot makes it easy to identify outliers, offering a clear indication of extreme values within the dataset.
- It summarizes key statistical information concisely, helping to understand how the data is distributed and where anomalies may exist.

Limitations:
- Unlike histograms, boxplots do not display the exact frequency of individual data points, which can be a limitation when granular insights are needed.
- For individuals unfamiliar with statistical concepts, boxplots may be more challenging to interpret, potentially leading to confusion if not explained properly.

Please see below for the Python code to generate grouped histograms for numeric columns, based on a categorical variable like room_type. This allows us to compare distributions across different categories.

I was not too sure how to create grouped histograms for a large population, so I googled "grouped histograms python numerical columns", which generated the following results:

```python
#draw grouped histograms on numerical columns 

#numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

#select categorical column for grouping (I used'room_type')
grouping_column = "room_type"

#filter out rows with missing grouping values
df = df.dropna(subset=[grouping_column])

#get different categories (I used 4 since there were only 4 room types. typically 3-5 can be used for ease of reading / understanding)
categories = df[grouping_column].value_counts().index[:4]  

#set figure size
plt.figure(figsize=(15, 12))

#loop through numeric columns and create grouped histograms. plot histograms for each category
for i, column in enumerate(numeric_columns[:6]):  
    plt.subplot(2, 3, i + 1) 
    for category in categories:
        subset = df[df[grouping_column] == category]
        sns.histplot(subset[column], bins=30, label=category, alpha=0.6, edgecolor="black", kde=True)

    #clean up graph -- labels, legend, layout 
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.title(f"Distribution of {column} by {grouping_column}")
    plt.legend(title=grouping_column)
plt.tight_layout()
plt.show()
```

<b>Answer:</b> Are there any outliers present?  (Data points that are far from the others.)

For this question, I will focus on price. 

Using the Interquartile Range (IQR) Method, we can identify that there are negative prices, which indicate errors. Also, 10.14% of all listings are considered outliers. See python code below. 

Key Findings:
- Total Listings Analyzed: 11,374
- Number of Outliers: 1,153
- Percentage of Outliers: 10.14% of all listings
- Lower Bound for Normal Prices: $-131 
- Upper Bound for Normal Prices: $445

```python
#convert 'price' column to numeric and drop NaN values
df["price"] = pd.to_numeric(df["price"], errors="coerce")
df_clean = df.dropna(subset=["price"])  # Remove NaN values

#define a function to detect outliers using the IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = data[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR  # Lower bound for outliers
    upper_bound = Q3 + 1.5 * IQR  # Upper bound for outliers

    #identify outliers
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    
    return outliers, lower_bound, upper_bound

#detect outliers in the 'price' column
outliers, lower_bound, upper_bound = detect_outliers_iqr(df_clean, "price")

#results
outlier_count = len(outliers)
total_entries = len(df_clean)
outlier_percentage = (outlier_count / total_entries) * 100
print(f"Total Listings Analyzed: {total_entries}")
print(f"Number of Outliers: {outlier_count}")
print(f"Percentage of Outliers: {round(outlier_percentage, 2)}%")
print(f"Lower Bound for Normal Prices: ${round(lower_bound, 2)}")
print(f"Upper Bound for Normal Prices: ${round(upper_bound, 2)}")
```
