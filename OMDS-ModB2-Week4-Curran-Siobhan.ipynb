Below are a histogram and boxplot representation of the same data. A pharmacy is keeping a record of the prices of the drugs that it sells, and an administrator wants to know how much the more expensive drugs tend to cost, in the context of the other prices.

Please write a short explanation of the pros and cons of these two representations. Which would you choose? How would you modify the formatting, if at all, to make it more visually interesting, clear, or informative?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(0)
num_data = 100
data = np.exp(np.random.uniform(size = num_data) * 4)
df = pd.DataFrame(data.T, columns = ["data"])
```

```python
print("The 75th percentile is:", df.quantile(q = 0.75))
df.plot.hist()
```

```python
df.plot.box()
```

Both histograms and boxplots have distinct advantages and limitations in representing data.

<b><i>Histograms - </i></b>

Advantages:
- Histograms offer a straightforward way to visualize the frequency of values within specified ranges, making it easy to quickly understand how data points are distributed.
- They provide a clear depiction of how the data is spread across different price ranges.
- Histograms help identify the concentration of data points, as well as the distribution's tails, which can be useful for spotting patterns and trends.

Limitations:
- The number of bins chosen for the histogram can influence interpretation. Too many bins may create unnecessary complexity, while too few can fail to capture the variation within the data.
- Histograms do not clearly highlight extreme outliers, which can be important for understanding the presence of unusually high or low values, especially in pricing data.

<b><i>Boxplot - </i></b>

Advantages:
- Boxplots deliver an efficient summary of the data, providing insight into the central tendency, spread, and skewness, including the median and quartiles.
- They are highly effective for comparing different sets of data, making them useful when analyzing multiple distributions side-by-side.
- The boxplot makes it easy to identify outliers, offering a clear indication of extreme values within the dataset.
- It summarizes key statistical information concisely, helping to understand how the data is distributed and where anomalies may exist.

Limitations:
- Unlike histograms, boxplots do not display the exact frequency of individual data points, which can be a limitation when granular insights are needed.
- For individuals unfamiliar with statistical concepts, boxplots may be more challenging to interpret, potentially leading to confusion if not explained properly.

From my professional experience, I have found that histograms are generally more intuitive for a broader audience. Having worked with data in financial risk analysis, I have observed that some individuals struggle with interpreting histograms. They often prefer more straightforward visualizations, such as bar charts, or rely on a detailed explanation from the presenter rather than analyzing the data themselves. Personally, I have yet to encounter a boxplot in professional settings, making them less familiar to many audiences.

The choice between a histogram and a boxplot ultimately depends on the audience's familiarity with statistical concepts. If the administrator has a strong understanding of statistics, I would recommend using the boxplot. It addresses their specific need to compare higher-priced drugs to others, clearly highlighting outliers (extreme prices) and providing a concise summary of the distribution, including the median and quartiles. The boxplot is especially suited for showing how more expensive drugs compare to the broader range of drug prices.

<b>Answer:</b>

I reproduced a  horizontal bar chart, as illustrated in Figure 2.23, page 64.

```python
#generate random data for drug price categories / corresponding average prices
np.random.seed(0)
categories = ['Painkiller', 'Antibiotic', 'Antidepressant', 'Antihistamine', 'Vaccine']
avg_prices = np.random.normal(loc=100, scale=20, size=len(categories))  # Random average prices for each category

#create horizontal bar chart
plt.figure(figsize=(10, 6))  # Size of the figure
plt.barh(categories, avg_prices, color='lightcoral', edgecolor='darkred')  # Horizontal bars

#add title, labels, gridlines
plt.title('Average Drug Prices by Category', fontsize=16, fontweight='bold', color='darkred')
plt.xlabel('Average Price ($)', fontsize=12)
plt.ylabel('Drug Categories', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
